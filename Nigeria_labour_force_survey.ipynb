```python
#read data
df, meta = pyreadstat.read_sav("/content/drive/MyDrive/NLFS_2024Q2_INDIVIDUAL.sav", encoding='latin1')
```

## Data cleaning and preprocessing

```python
print(df.shape)
df.head()
```

```python
## select columns relating to primary or main job
columns = ['id5_sector', 'dc3','dc5', 'dc6', 'ed7', 'mjj2ccleanmaingroup', 'mjj3a', 'mjj4', 'mjj8a', 'mjj8b_1', 'mjj8b_2', 'mjj8b_3', 'mjj8b_4',
          'mjj8b_5', 'mjj8b_6', 'mjj8b_7', 'mjj8b_8', 'mjj8b_9', 'mjj8c', 'mjj8g', 'mjj8k', 'mjj8i_1', 'mjj8i_2', 'mjj8i_97', 'mjj8m',
           'mjj8n', 'mjj9a', 'mjj9b', 'mjj9c', 'mjj12']

col_drop = [col for col in df.columns if col not in columns]

work_df = df.drop(columns = col_drop)
```

```python
# check missing values
work_df.isna().sum()
```

```python
work_df.describe()
```

# Visualize missing value and missingness type using missingno

```python
msno.matrix(work_df)
```

```python
# since the objective is analysis labour force, only valid responses was used
#drop row by individuals who don't have a primary job, no recorded job sector and no response on contract/agreement and specificed contract_duration
# and numbers number of work hours per week

drop_df = work_df.copy('deep')
drop_df.dropna(subset = ['mjj2ccleanmaingroup', 'mjj8a', 'mjj8g','mjj8c', 'mjj12'], inplace=True)

#drop rows where respnse is 'i don't know'
columns_to_check = ['mjj8m', 'mjj8n', 'mjj8k','mjj9a', 'mjj9c']
for col in columns_to_check:
    drop_df = drop_df[drop_df[col] != 97]

print(drop_df.shape)
```

```python
msno.matrix(drop_df)
```

```python
msno.heatmap(drop_df, figsize=(10, 8), cmap='viridis')
plt.xticks(rotation = 90)
plt.tick_params(axis = 'both', direction = 'inout')
plt.show()
```

```python
# merge response on payment method into one column
# Define conditions
conditions = [
    drop_df['mjj8b_1'] == 1,
    drop_df['mjj8b_2'] == 1,
    drop_df['mjj8b_3'] == 1,
    drop_df['mjj8b_4'] == 1,
    drop_df['mjj8b_5'] == 1,
    drop_df['mjj8b_6'] == 1,
    drop_df['mjj8b_7'] == 1,
    drop_df['mjj8b_8'] == 1,
    drop_df['mjj8b_9'] == 1
]

# Define corresponding choices
choices = [
    'wage',
    'piece of work completed',
    'commissions',
    'tips',
    'fees for services provided',
    'payment in meals or accommodation',
    'payment in products',
    'other cash payment',
    'not paid'
]

# Assign new column
drop_df['pay_method'] = np.select(conditions, choices, default='unknown')

#fill missing values
drop_df['dc6'] = drop_df['dc6'].fillna(drop_df['dc6'].mode()[0])
drop_df['ed7'] = drop_df['ed7'].fillna(drop_df['ed7'].mode()[0])
drop_df['dc5'] = drop_df['dc5'].fillna(drop_df['dc5'].mode()[0]) # they are all categorical variabls
#drop merged columns
main_df = drop_df.drop(columns = ['mjj8i_1', 'mjj8i_2', 'mjj8b_1', 'mjj8b_2', 'mjj8b_3', 'mjj8b_4',
                                  'mjj8b_5', 'mjj8b_6', 'mjj8b_7', 'mjj8b_8', 'mjj8b_9'])
```

```python
#rename columns
main_df.rename(columns={
    'id5_sector': 'sector','dc3': 'gender', 'dc5':'age', 'dc6':'marital_status', 'ed7':'edu_level', 'mjj2ccleanmaingroup':'main_job_cart',
    'mjj3a':'main_activity', 'mjj4':'employment_type', 'mjj8a': 'job_sector', 'mjj8c': 'contract_agreement', 'mjj8g': 'contract_duration',
    'mjj8k': 'tax_deduc', 'mjj8m': 'annual_leave','mjj8n':'sick_leave', 'mjj9a': 'work_environ', 'mjj9b': 'no_of_workers',
    'mjj9c': 'is_bus_reg', 'mjj12':'hours_per_week'
}, inplace=True)

main_df.reset_index(drop=True, inplace=True)
```

```python
# replace response integers with the actual response
main_df['sector'] = main_df['sector'].replace({
    1.0: 'urban',
    2.0: 'rural'
})
main_df['gender'] = main_df['gender'].replace({1.0:"male", 2.0:"female"})

main_df['marital_status'] = main_df['marital_status'].replace({
    1.0:'married_mono',
    2.0: 'married_poly',
    3.0: 'informal_losse_union',
    4.0: 'divorced',
    5.0: 'separated',
    6.0: 'widowed',
    7.0: 'never_married'
})

main_df['edu_level'] = main_df['edu_level'].replace({
    1.0: 'none',
    2.0: 'first_school_leaving_cert',
    3.0: 'modern_school_leaving_cert',
    5.0: 'junior_sec_school_cert',
    6.0: 'senior_sec_school_cert',
    7.0: 'advanced_level',
    8.0: 'nc_nd_nursing',
    9.0: 'ba_bsc_hnd',
    10.0: 'tech_prof',
    11.0: 'masters',
    12.0: 'doctorate',
    41.0: 'voc_comm_cert',
    42.0: 'voc_comm_diploma',
    96.0: 'others'
})

main_df['main_job_cart'] = main_df['main_job_cart'].replace({
    0.0: 'armed forces',
    1.0 : 'managers',
    2.0: 'professionals',
    3.0: 'technicians and associate professionals',
    4.0:  'clerical support workers',
    5.0: 'services and sales workers',
    6.0:  'skilled agricultural forestry and fishery workers',
    7.0: 'craft and related trades workers',
    8.0 : 'plant and machine operators and assemblers',
    9.0: 'elementary occupations'
})

main_df["employment_type"] = main_df["employment_type"].replace({
    1.0: 'employee',
    2.0: 'own business/farming activity',
    3.0: 'helping in a household business',
    4.0: 'apprentice/intern',
    5.0: 'helping a household member who works for someone else'
})

main_df['job_sector'] = main_df['job_sector'].replace({
    1.0: 'federal government',
    2.0: 'state government',
    3.0: 'local government',
    4.0: 'state-owned enterprise',
    5.0: 'private sector',
    6.0: 'farm',
    7.0: 'domestic worker',
    8.0: 'NGO',
    9.0: 'co-operatives',
    10.0: 'international organization',
    11.0: 'religious organization',
    12.0: 'others'
})

main_df['contract_agreement'] = main_df['contract_agreement'].replace({
    1.0: 'written contract',
    2.0: 'oral agreement'
})

main_df['contract_duration'] = main_df['contract_duration'].replace({
    1.0: 'for a specified period of time',
    2.0: 'until the date a task is completed',
    3.0: 'permanent or until retirement',
    4.0: 'no specified end date'
})

main_df['tax_deduc'] = main_df['tax_deduc'].replace({
    1.0: 'employer is responsible',
    2.0: 'self responsible',
    3.0: 'not applicable'
})

main_df['annual_leave'] = main_df['annual_leave'].replace({
    1.0: 'yes',
    2.0: 'no'
})

main_df['sick_leave'] = main_df['annual_leave'].replace({
     1.0: 'yes',
    2.0: 'no'
})

main_df['work_environ'] = main_df['work_environ'].replace({
    1.0: 'my home',
    2.0: 'employers home',
    3.0: 'farm',
    4.0: 'fixed premise',
    5.0: 'public space',
    6.0: 'in or on a vehicle',
    7.0: 'door to door',
    8.0: 'others'
})

main_df['no_of_workers'] = main_df['no_of_workers'].replace({
    1.0: '0-1',
    2.0: '2-4',
    3.0: '5-9',
    4.0: '10-19',
    5.0: '20-49',
    6.0: '50+',
})
main_df['is_bus_reg'] = main_df['is_bus_reg'].replace({
     1.0: 'yes',
     2.0: 'no'
})
main_df.to_csv('Labour_Survey.csv',index=False)

# HOURS SPENT AT WORK PER WEEK
bins = [0, 14, 29, 34, 44, 59, 79, 99, 119, 139, 168]
labels = [
    '0-14',
    '15-29',
    '30-34',
    '35-44',
    '45-59',
    '60-79',
    '80-99',
    '100-119',
    '120-139',
    '140-168'
]

# Create a new column for categorized weekly hours
main_df['hours_per_week'] = pd.cut(main_df['hours_per_week'], bins=bins, labels=labels)

#convert data types
main_df['age']  = main_df['age'].astype(int)
# Define age bins and labels
age_bins = [9, 18, 28, 38, 48, 58, 68, 78, 88, float('inf')]
age_labels = [
    '9-18',
    '19-28',
    '29-38',
    '39-48',
    '49-58',
    '59-68',
    '69-78',
    '79-88',
    '89+'
]

# Categorize the ages
main_df['age'] = pd.cut(main_df['age'], bins=age_bins, labels=age_labels)

main_df['main_activity'] = main_df['main_activity'].str.lower() #conver to lower case

main_df.info()
```

## Data Visualization

```python
sns.histplot(data=main_df, x='age', kde=False, bins=10)
plt.title('Workforce age distribution')
plt.show()
plt.savefig('age_distribution.png')
```

```python
fig = plt.figure(figsize=(10,15))
sns.catplot(data=main_df, kind = 'count', y = 'pay_method',order= main_df['pay_method'].value_counts().index, aspect=1.5)
plt.xticks(rotation=90)
plt.title("Payment method distribution")
plt.xlabel("Count")
plt.ylabel('payment method')
plt.show()
```

```python
fig = plt.figure(figsize=(10,15))
sns.catplot(data=main_df, kind = 'count', y= 'edu_level', order = main_df['edu_level'].value_counts().index, aspect=1.5)
plt.xticks(rotation=90)
plt.title("Highest Education Qualification")
plt.xlabel("Count")
plt.ylabel('Education level')
plt.show();
```

Insight: The Typical full-time work hours per week falls between 35-44 hours

```python
fig = px.bar(main_df, x='edu_level', color = main_df['hours_per_week'].sort_values(), opacity =1, title = 'Education qualification VS Hours per week')
fig.update_layout(
    plot_bgcolor='darkslategray',
    paper_bgcolor='blanchedalmond',
    barmode = 'stack'
)
```

Insight: There are more females in the professional fields than in the army,  manufactiring factories or agriculture.

```python
fig = plt.figure(figsize=(10,15))
sns.catplot(data=main_df, kind = 'count', y= 'main_job_cart', hue = 'gender', order = main_df['main_job_cart'].value_counts().index, aspect=1.9)
plt.xticks(rotation=90)
plt.title("Gender distribution vs Job categories")
plt.xlabel("Count")
plt.ylabel('Job categories')
plt.show()
plt.savefig('gender_dist_job_categories.png')
```

Insights: Individuals who are unmarried tends to work longer hours per week.

Insight: Are people more likely to get permanent jobs in public or private sector?
